### Histogram for FIFA Data


```python
# Loading data in 'fifa' dataframe
fifa = pd.read_csv("fifa_data.csv", index_col = 0)
fifa
```

```python
plt.figure(figsize = (9,6))

# Bars of Histogram are called bins
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins = bins , color ="lightblue")

plt.title("Distribution of Players Skills in FIFA 2018")
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")

plt.savefig("Images/FIFA_Histogram.png", dpi=300)
plt.show()
```

### Pie-Chart for FIFA Data

```python
# Analyse all columns of DataFrame
fifa.columns
```

```python
# Pie Chart for showing how many players opt right foot for play and how many opt left foot to play
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]   # count() wil output array, count()[0] will show value
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0] 
```

```python
plt.figure(figsize = (8,5))
plt.pie([left, right], labels = ['Left', 'Right'], colors =['Aqua','Teal'], autopct = '%.2f%%')
plt.title("Foot Preference for FIFA Players")
plt.savefig("Images/FIFA_Pie-Chart01.png", dpi = 200)
plt.show()
```

```python
fifa.Weight
```

```python
#Stripping off "lbs" from weights to perform functions on numerical values of weight
def utility_func(rows):
        if type(rows.Weight)==str:
            return int(rows.Weight[:len(rows.Weight)-3])  # Using string slicing technique to get slice "lbs"  
            # return int(rows.Weight.strip("lbs"))

fifa.Weight = fifa.apply(utility_func,axis=1)
fifa.Weight  # Weights in numerical values
```

```python
# Making categories in weights
light = fifa.loc[fifa['Weight'] < 125].count()[0]
light_medium = fifa.loc[(fifa['Weight'] >= 125) & fifa.Weight < 150 ].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]
```

```python
# Plotting Pie-Chart
plt.figure(figsize=(8,5))

# Different color schemes of pie-chart
#plt.style.use("default")
plt.style.use("ggplot")

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Light', 'Light_Medium', 'Medium', 'Medium_Heavy', 'Heavy']

plt.pie(weights, labels =labels, autopct ='%.2f%%')
plt.title('Weight of Professional Soccer Players (lbs)')
plt.savefig("Images/FIFA_Pie-Chart02.png", dpi=200)
plt.show()
```

```python
# Fixing the noisy pie-chart
plt.figure(figsize=(8,5))
explode = (0.5, 0.1, 0, 0, 0.9)    # 0.5 and 0.9 are for Light and Heavy portion, since they were squashed together
plt.pie(weights, labels =labels, autopct ='%.2f%%', explode= explode)
plt.title('Weight of Professional Soccer Players (lbs)')
plt.savefig("Images/FIFA_Pie-Chart03.png", dpi=200)
plt.show()
```

### Box and Whiskers Plot

```python
# Box and Whiskers are good to visualize comparsions in dataset attributes

plt.figure(figsize =(5,8))   #Vertical box-plots gives good visualization
plt.style.use("default")

barcelona = fifa.loc[fifa.Club =='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

boxes = plt.boxplot([barcelona,madrid, revs], labels= ['FC Barcelona','Real Madrid', 'NE Revolution'], patch_artist = True)
plt.title("Professional Soccer Team Players Comparison")
plt.ylabel("FIFA Overall Performance")

# Another way to set properties. In order to use facecolor property, first set patch_artist = True in attribute
for box in boxes['boxes']:
    box.set(color = "blue", linewidth= 3, hatch= '/', facecolor= "orange")
plt.savefig("Images\FIFA_Box-Whiskers.png", dpi= 100)
plt.show()
```
