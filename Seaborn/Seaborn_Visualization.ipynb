```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
height = [10,20,30,34,40,45,55,67]
weight = [23,33,44,45,55,67,70,79]
```

```python
sns.scatterplot(x= height, y=weight)
```

### Countplot
Countplot takes categorical observations and plot bar graph of the distinct categories from the observations.

```python
gender = ["Female", "Male", "Male", "Male", "Female"]
sns.countplot(x=gender)
```

```python
# To use it with dataframe
fifa = pd.read_csv("fifa_data.csv", index_col = 0)
sns.countplot(x= "Nationality", data= fifa)
```

### Hues

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# adding hue (subgroups)
sns.scatterplot(x="total_bill", y="tip", data=tips, hue= "smoker")
```

```python
# adding palette
hue_color = {'Yes':'black', "No": 'red'}
sns.scatterplot(x="total_bill", y="tip", data=tips, hue= "smoker", palette = hue_color)
```

## Visualizing Two Quantitative Variables

Scatter plots are relational plots. Relplots() are used for quantitative variables visualization either by scatter plots or line plots.

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue= "smoker", kind= "scatter", palette = hue_color)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind= "scatter", col="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind= "scatter", row="smoker")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind= "scatter",row="time", col="smoker" )
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind= "scatter", col="day", col_wrap = 3 ) # ek col me 3 instance hu
```

```python
# Customizing sns relplot with size , style and alpha transparency
sns.relplot(x="total_bill", y="tip", data=tips, kind= "scatter", size="size", hue="size", style="size" ,alpha=0.4)
```

```python
# Line plots (Measure change over time between two variables)
sns.relplot(x="tip", y="total_bill", data=tips , kind="line", hue="sex", style="sex", markers=True)
```

## Visulizing Categorical and Quantitative Data
Countplot and Barplots are for categorical data

```python
sns.catplot(x="size", data=tips, kind="count" )   #order parameter is used to order the categories
```

#### Box Plots
Quantitative data on categorical groups on x-axis 

```python
sns.catplot(x="time", y="total_bill",  data= tips, kind="box")  #line shows distribution, diamonds shows outliers, 
# line inside box shows median
```

```python
sns.catplot(x="time", y="total_bill",  data= tips, kind="box", order=["Dinner", "Lunch"], sym="")  # sym removes outliers
```

#### Point Plots

```python
sns.catplot(x="time", y="total_bill",  data= tips, kind="point")
```

```python
sns.catplot(x="time", y="total_bill",  data= tips, kind="point", join=False, capsize=0.2)
```

```python
# Assigning Titles
g = sns.catplot(x="time", y="total_bill",  data= tips, kind="point", join=False, capsize=0.2)
g.fig.suptitle("Insert title", y=1.03)  # Y is distnace from title to graph
g.set(xlabel= "New X label", ylabel= "New Y label")
plt.xticks(rotation = 90)
```
